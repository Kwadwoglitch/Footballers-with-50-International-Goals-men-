```python
sports_dataFrame = pd.read_csv("dataset.csv", encoding="latin1")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
numeric_data = sports_dataFrame.select_dtypes(['int64','float64']) 
corr_matrix = numeric_data.corr()
sns.heatmap(corr_matrix, annot=True)
plt.title('CORRELATION BETWEEN RANKS, GOALS, CAPS & GOALS PER MATCH')
plt.show()
```

```python
Top_10_players = sports_dataFrame.nlargest(10, 'Goals')
sns.barplot(data=Top_10_players,x='Player',y='Goals',hue='Nation', palette='Set2')
plt.title('TOP 10 PLAYERS BY GOALS')
plt.xticks(rotation=-45)
plt.legend(loc=(1.1,0.3))
plt.show()
```

```python
nation_goals = sports_dataFrame.groupby('Nation')['Goals'].mean().sort_values()
plt.figure(figsize=(6,10))
nation_goals.plot(kind='barh',title='GOALS DISTRIBUTION BY NATION')
plt.xticks(rotation=90)
plt.show()
```

```python
avg_goals = sports_dataFrame.groupby('Confederation')['Goals'].mean().sort_values()
avg_goals.plot(kind='bar',title='AVERAGE GOALS BY CONFEDERATION')
plt.xticks(rotation=-45)
plt.show()
```

```python
confederation_goals = sports_dataFrame.groupby('Confederation')['Goals'].sum().reset_index()

plt.pie(confederation_goals['Goals'], labels=confederation_goals['Confederation'],
    autopct='%1.1f%%', startangle=90, colors=sns.color_palette("pastel"))

# Add a title
plt.title("Goals Distribution by Confederation")
plt.show()
```

```python
sns.scatterplot(sports_dataFrame, x='Caps', y='Goals', hue='Confederation')
plt.title('GOALS VS CAPS BY CONFEDERATION')
plt.legend(loc=(1.1,0.5))
plt.show()
```

```python
sports_dataFrame[['Career Start', 'Career End']] = sports_dataFrame['Career span'].str.split('-', expand=True)


sports_dataFrame['Career Start'] = pd.to_numeric(sports_dataFrame['Career Start'])
sports_dataFrame['Career End'] = pd.to_numeric(sports_dataFrame['Career End'])

sports_dataFrame['Career Duration'] = sports_dataFrame['Career End'] - sports_dataFrame['Career Start']

sns.boxplot(
    data=sports_dataFrame,
    x='Confederation',
    y='Career Duration',
    palette="pastel"
)

# Add title and labels
plt.title("Career Duration by Confederation")
plt.xlabel("Confederation")
plt.ylabel("Years Active")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.scatterplot(sports_dataFrame, x='Goals per match', y='Goals', hue='Confederation')
plt.title('Goals per Match vs. Goals Scored')
plt.legend(loc=(1.1,0.5))
plt.show()
```
